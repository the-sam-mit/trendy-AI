```python
%%writefile app.py

import streamlit as st
import pandas as pd

# Set page to wide mode for a professional dashboard look
st.set_page_config(page_title="TrendPilot AI", layout="wide")

# Custom CSS for styling
st.markdown("""
    <style>
    .metric-container { background-color: #f0f2f6; padding: 20px; border-radius: 10px; }
    </style>
""", unsafe_allow_html=True)

# Sidebar for platform selection and settings
with st.sidebar:
    st.title("âš™ï¸ TrendPilot Settings")
    platform = st.selectbox("Target Platform", ["Reddit", "X (Twitter)", "TikTok"])
    target_audience = st.text_input("Target Audience", "US Youth / Gen-Z")
    if st.button("ğŸ”„ Sync Live Data"):
        st.success("Fetched latest profile metrics!")

# Main Title
st.title("ğŸš€ TrendPilot AI: Content Strategy Dashboard")
st.divider()

# ROW 1: On-the-Fly Creator Metrics
st.subheader("ğŸ“Š Your Live Performance (On-the-Fly)")
col1, col2, col3, col4 = st.columns(4)
col1.metric("Avg. Engagement", "8.5%", "+1.2%")
col2.metric("Best Post Time", "10:00 AM", "EST")
col3.metric("Follower Growth", "1,240", "+12%")
col4.metric("Retention Rate", "64%", "-2%")

st.divider()

# ROW 2: Two-column layout for RAG Trends vs AI Suggestions
col_left, col_right = st.columns([1, 1.5])

with col_left:
    st.subheader("ğŸ”¥ Current Trends (RAG)")
    trends = [
        {"Topic": "AI Art Challenges", "Platform": "Reddit", "Heat": 98},
        {"Topic": "Sustainable Fashion", "Platform": "X", "Heat": 85},
        {"Topic": "Retro Gaming Tech", "Platform": "Reddit", "Heat": 72}
    ]
    st.table(pd.DataFrame(trends))
    st.info("ğŸ’¡ Data refreshed 14 mins ago via Vector DB.")

with col_right:
    st.subheader("ğŸ¤– AI Strategic Suggestion")
    with st.container():
        st.markdown(f"""
        **The Analysis:** Your audience on **{platform}** is currently engaging most with *retro-style* content,
        which overlaps with the current **'Retro Gaming'** trend.

        **Recommended Post:**
        - **Content Type:** 'How-to' carousel or short-form video.
        - **Hook:** "Why [Trend Topic] is changing the way we see {target_audience} culture."
        - **Best Time to Post:** **Today at 4:15 PM EST** (Based on your peak reach).
        """)
        if st.button("ğŸ“… Add to Schedule"):
            st.balloons()
```
