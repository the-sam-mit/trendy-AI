**Accessing Gemini LLM using ChatGoogleGenerativeAI**

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
responseLLM = llm.invoke("What is streamLit?")

print(responseLLM.content)
```
