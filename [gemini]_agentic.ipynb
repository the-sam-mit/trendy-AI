```python
import os
from crewai import LLM

# Ensure your GEMINI_API_KEY is set in Colab Secrets
from google.colab import userdata
gemini_api_key = userdata.get("GEMINI_API_KEY")

# If you prefer to set it directly (not recommended for sensitive keys):
# os.environ["GEMINI_API_KEY"] = "YOUR_GEMINI_API_KEY" # Replace with your actual key

llm = LLM(model="gemini/gemini-2.5-flash-lite", api_key=gemini_api_key, temperature=0.0)
```

```python
# import streamlit as st
from crewai import Agent, Task, Crew, Process
from crewai.tools import tool
from langchain_community.tools import DuckDuckGoSearchRun

# Setup Search Tool
ddg_search = DuckDuckGoSearchRun()

@tool('DuckDuckGoSearch')
def search_tool(search_query: str):
  """Search tool function for search query"""
  return ddg_search.run(search_query)

# st.title("🚀 AI Content Creator Crew")
# st.markdown("Enter a platform to generate trending content.")

# platform = st.text_input("Social Media Platform", "Twitter/X")
# topic_area = st.text_input("Niche/Topic Area", "Artificial Intelligence")
platform = "Twitter/X"
topic_area = "Artificial Intelligence"

# if st.button("Generate Content"):
#     with st.spinner('Agents are working...'):

# Agent 1: The Trend Researcher
researcher = Agent(
    role='Market Researcher',
    goal=f'Identify trending topics on {platform} about {topic_area}',
    backstory='Expert at spotting viral trends and understanding audience engagement.',
    tools=[search_tool],
    llm=llm,
    verbose=True
)

# Agent 2: The Content Reviewer
reviewer = Agent(
    role='Content Quality Lead',
    goal='Review content for virality, tone, and engagement. Provide harsh feedback if it lacks quality.',
    backstory='A seasoned social media manager who knows what gets clicks and what fails.',
    llm=llm,
    verbose=True
)

# Task 1: Research and Draft
task_research = Task(
    description=f'Find 3 trending topics for {platform} in {topic_area} and draft a post for each.',
    expected_output='A list of 3 drafted posts based on current trends.',
    agent=researcher
)

# Task 2: Review and Refine
task_review = Task(
    description='Review the drafted posts. If they are not engaging, rewrite them. Ensure they include relevant hashtags.',
    expected_output='A final polished version of the social media posts.',
    agent=reviewer
)

# Define the Crew
crew = Crew(
    agents=[researcher, reviewer],
    tasks=[task_research, task_review],
    process=Process.sequential, # Reviewer follows Researcher
    tracing = True
)

result = crew.kickoff()
```
